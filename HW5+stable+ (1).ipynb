```julia
#remember that the cn and xn of n=2 is 1,1, +-sqrt3/3  
using QuadGK
using Calculus
g(x)=4/(1+x^2)
Ic=integrate(g,0,1)
#Ic=pi
```

```julia
function gauss(h,a,b)
    f(t)=4/(1+(((b-a)*t+a+b)/2)^2)
    t1=-sqrt(15)/5
    t2=0
    t3=sqrt(15)/5
    I1=( 5/9*f(t1)+8/9*f(t2) +5/9*f(t3))*(b-a)/2
    return I1
end
```

```julia
function summedgauss(k)
    h=(1-0)/k
    I=0
    for i=1:k
        a=0+(i-1)*h
        b=0+(i)*h
        I=I+gauss(h,a,b)
    end
    return I
end
```

```julia
phi(x)=(x^2)*log(x)
Ir=integrate(phi,1,2)
```

```julia
function summedtreprzoid(n,a,b)
    h=1/2^(n)
    sum=(phi(a)+phi(b))*h/2
    for i=1:2^n-1
        sum=sum+h*phi(i*h+a)
    end
    return sum
end
```

```julia
function Romberg(a,b)
    R=zeros(4,4)
    for i=1:size(R)[1]
        R[i]= summedtreprzoid(i-1,a,b)
    end
    for j=2:size(R)[1]
        for k=j:size(R)[1]
            R[k,j]=R[k,j-1]+((R[k,j-1]-R[k-1,j-1])/((2^(j-1))^2-1))
        end
    end
    return R
end
```

```julia
a=1
b=2
R=Romberg(a,b)
```

```julia
function errorsummedtreprzoid(n,a,b)
    h=(b-a)/(n+1)
    sum=(phi(a)+phi(b))*h/2
    for i=1:n
        sum=sum+h*phi(i*h+a)
    end
    return sum-Ir
end
```

```julia
function summedtreprzoidold(n,a,b,error1,error)   
    while(abs(error1)>error)  
    error1=errorsummedtreprzoid(n,a,b)
    n=n+1
    end
println("We need ",n," (not include a,b) points to reach the ", error ," precision by In(1)(f)")
end
```

```julia
function newton(F, dF, x0, precision = 1.e-8)
    x = x0;
    for i = 1:100
        step = - inv(dF(x)) * F(x)
        x = x + step
        if (norm(F(x)) < precision)
            return x
        end
    end
    error("no convergence")
end
```

```julia
function backward_euler(f, df, t0, y0, h, T)
    d = length(y0)
    N = convert(Int64, floor(T/h))
    
    t = zeros(1, N + 1)
    y = zeros(d, N + 1)

    # Initial values:
    t[1] = t0
    y[:,1] = y0

    for i = 1:N
        t[i+1] = t[i] + h
        
        G(z) = y[:,i] + h * f(t[i+1], z) - z
        dG(z) = h * df(t[i+1], z) - eye(d)
        y[:,i+1] = newton(G, dG, y[:,i])
    end
    
    return (t, y)
end
```

```julia
f(t,x)=[
    -x[2]-x[3]
    x[1]+0.1*x[2]
    0.1+x[3]*(x[1]-14)
]

df(t,x)=[
    0 -1 -1
    1 0.1 0
    x[3] 0 x[1]-14
]
```

```julia
t0 = 0.
y0 = [ 10.; 10.;0. ]
h=0.001
T=40;
```

```julia
t,y=backward_euler(f, df, t0, y0, h, T)
y
```
